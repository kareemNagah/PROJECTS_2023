## __cleaning data__

```python
# drop id column because it is not useful 
df.drop( columns = "ID" , inplace = True )
```

```python
# drop null values in Arrival Delay column
df.dropna( inplace = True) # drop null values
```

```python
# checking outliers in the dataset :
df.hist(bins=30,figsize=(15,15))
plt.show()
```

>### numerical columns outliers check :


```python
num_cols = ['Age','Flight Distance','Departure Delay','Arrival Delay']
for col in num_cols:
    sns.catplot(x=col,data=df,kind='box')
    plt.show()
```

>>### _so as we can see there is no outliers in ['Flight Distance','Departure Delay','Arrival Delay'] columns_

```python
# removing outliers in the dataset :
outliers = ['Flight Distance','Departure Delay','Arrival Delay']
for col in outliers :
    q1 = df[col].quantile(0.25) # first quartile value 
    q3 = df[col].quantile(0.75) # third quartile value 
    iqr = q3 - q1 # interquartile range 
    lower_bound = q1 - (1.5 * iqr) # lower bound 
    upper_bound = q3 + (1.5 * iqr) # upper bound 
    df.loc[(df[col] < lower_bound) | (df[col] > upper_bound),col] = df[col].median() # replacing outliers with median value 
```

```python
# checking outliers in the dataset :
df.hist(bins=30,figsize=(15,15))
plt.show()
```
